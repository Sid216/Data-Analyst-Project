```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
```

```python
df = pd.read_csv('E:\Data Analyst Project\python\kc_house_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.drop(['id'], axis=1, inplace=True)
df.describe()
```

```python
#check for missing values in the bedroom and bathroom columns

print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
#replace missing values in bedrooms with mean value of bedrooms

mean=df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)

#doing the same for bathrooms

mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean, inplace=True)
```

```python
print("Number of NaN values for the column bedrooms: ", df['bedrooms'].isnull().sum())
print("Number of NaN values for the column bathrooms: ", df['bathrooms'].isnull().sum())
```

```python
#creating a boxplot to show relation and outliers between houses with a waterfront view and high prices
df["waterfront"] = df["waterfront"].map({0:'No', 1:'Yes'})
sns.boxplot(x='waterfront', y='price', data=df)
plt.xlabel("Waterfront View")
plt.ylabel("Price")
plt.title("Waterfront vs Prices")
```

```python
#using regplot to determine relation of sqft_above and price
sns.regplot(x='sqft_above', y='price', data=df, line_kws={'color':'red'})
```

```python
#linear regression model using longitude and calculate R^2
x = df[['long']]
y = df['price']
lm = LinearRegression()
lm.fit(x,y)
lm.score(x,y)
```

```python
#linear regression model using sqft_living and calculate R^2
x = df[['sqft_living']]
y = df['price']
lm = LinearRegression()
lm.fit(x,y)
lm.score(x,y)
```

```python
#linear regression model to predict price using these features
features =["floors", "lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
x = df[features]
y = df['price']
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=1)

# Create the linear regression model
lm = LinearRegression()

# Fit the model using the training data
lm.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = lm.predict(X_test)

# Evaluate the model
r2 = r2_score(y_test, y_pred)

print(f"R^2 Score: {r2}")
```

```python
#using the list to create a pipeline object to predict the 'price', fit the object using the features in the list features, and calculate the R^2.

Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(x,y)
pipe.score(x,y)
```

```python
pr=PolynomialFeatures(degree=2)
```

```python
#
x_train_pr=pr.fit_transform(x_train[['floors', 'lat' ,'bedrooms' ,'sqft_basement' ,'view' ,'bathrooms','sqft_living15','sqft_above','grade','sqft_living']])
x_polly=pr.fit_transform(x_train[['floors', 'lat' ,'bedrooms' ,'sqft_basement' ,'view' ,'bathrooms','sqft_living15','sqft_above','grade','sqft_living']])

RidgeModel=Ridge(alpha=0.1)
RidgeModel.fit(x_train_pr, y_train)
RidgeModel.score(x_train_pr, y_train)
```
